```python
from sentence_transformers import SentenceTransformer

# Load a sentence transformer model (MiniLM works well)
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Convert news descriptions to embeddings
descriptions = [
    "Market hits record highs.",
    "Bitcoin reaches $100K.",
    "AI companies lead the way."
]
embeddings = model.encode(descriptions)

# Store embeddings in FAISS
import faiss
import numpy as np

index = faiss.IndexFlatL2(384)  # 384-dimension vectors for MiniLM
index.add(np.array(embeddings))  # Add vectors to FAISS
```

```python
def search_news(query, top_k=2):
    # Convert query to embedding
    query_embedding = model.encode([query])

    # Search in FAISS
    distances, indices = index.search(np.array(query_embedding), top_k)

    # Return top results
    results = [descriptions[i] for i in indices[0]]
    return results

# Example query
query = "Tell me about the stock market."
retrieved_news = search_news(query)
print("Retrieved News:", retrieved_news)
```

```python
from transformers import pipeline

# Load a generative model (GPT-2 or similar)
rag_model = pipeline("text-generation", model="gpt2")

def generate_response(query):
    retrieved_texts = search_news(query)
    context = " ".join(retrieved_texts)  # Combine relevant articles

    # Use the generative model to generate an answer
    prompt = f"Context: {context}\n\nUser Query: {query}\n\nAnswer:"
    response = rag_model(prompt, max_length=100, do_sample=True)[0]['generated_text']

    return response

# Example RAG response
query = "What's happening in crypto?"
response = generate_response(query)
print("Generated Answer:", response)
```

```python
from fastapi import FastAPI
from pydantic import BaseModel
from transformers import pipeline
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Initialize FastAPI
app = FastAPI()

# Load sentence transformer for embeddings
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Sample news data
descriptions = [
    "Market hits record highs.",
    "Bitcoin reaches $100K.",
    "AI companies lead the way."
]
embeddings = model.encode(descriptions)

# Initialize FAISS
index = faiss.IndexFlatL2(384)  # 384-dimension vectors for MiniLM
index.add(np.array(embeddings))

# Load text generation model
rag_model = pipeline("text-generation", model="gpt2")

# Request model
class QueryRequest(BaseModel):
    query: str

# Function to search news
def search_news(query, top_k=2):
    query_embedding = model.encode([query])
    distances, indices = index.search(np.array(query_embedding), top_k)
    results = [descriptions[i] for i in indices[0]]
    return results

# Function to generate response
def generate_response(query):
    retrieved_texts = search_news(query)
    context = " ".join(retrieved_texts)
    prompt = f"Context: {context}\n\nUser Query: {query}\n\nAnswer:"
    response = rag_model(prompt, max_length=100, do_sample=True)[0]['generated_text']
    return response

# API Endpoint
@app.get("/")
def home():
    return {"message": "RAG API is running! Use POST /rag to query."}

@app.post("/rag")
def rag_endpoint(request: QueryRequest):
    return {"response": generate_response(request.query)}

# Run using: uvicorn filename:app --host 0.0.0.0 --port 8000
```
